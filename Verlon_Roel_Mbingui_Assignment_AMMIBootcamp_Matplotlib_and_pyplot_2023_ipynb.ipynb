```python
# data for plottig the figure
x_axis = ["State SAC", "CURL", "Dreamer", "SLACv1", "SAC+AE", "PlaNet", "Pixel SAC"]
env_steps_100k = [779, 560, 296, 437, 351, 137, 133] # Values for the blue bars
env_steps_500k = [923, 914, 795, 767, 745, 406, 188] # Values for the red bars
y_label = "Evaluation Score" 
title = "Median Scores on DMControl100k and DMControl500k"
```

```python
############## Add your code here ##############
'''
General Steps:
1. Make the multibar plot with the required labels, colors, and legend.
2. Make sure the font size is 20.
3. Add the title and the other stylistic modifications.
'''
index = np.arange(len(x_axis)) 
width = 0.35 
plt.bar(index-0.2, env_steps_100k, width=0.35, label='env_steps_100k', color = 'b')
plt.bar(index + 0.2, env_steps_500k, width=0.35,label='env_steps_500k', color ='r')
for i, v in enumerate(env_steps_100k):
    plt.text(i - width/2, v + 10, str(v), ha='center', va='bottom', fontsize=20)
for i, v in enumerate(env_steps_500k):
    plt.text(i + width/2, v + 10, str(v), ha='center', va='bottom',fontsize=20)
    
plt.ylim(0,1000)
plt.xticks(index + width // 2, x_axis, fontsize=20)
plt.legend(fontsize=20)
plt.ylabel("Evaluation Score",fontsize=20)
plt.title("Median Scores on DMControl100k and DMControl500k",fontsize=20)
################################################
plt.show()
```
